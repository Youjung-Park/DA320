```python
print("DA320 Assignment2 - Jupyter Notebooks")
print("Youjung Park")
print("Date : 04/17/2023")
print()

# import urllib3 to make http requests
import urllib3

# import Regex to write regular expressions
import re

# this is the url that I am going to contact. (I chose movie year 2019 from metacritic.)
url = 'https://www.metacritic.com/browse/movies/score/metascore/year/filtered?year_selected=2019&sort=desc&view=detailed'

# construct PoolManager to generate a request
http = urllib3.PoolManager()

# use User-Agent to fix redirect issues 
header = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/34.0.1847.131 Safari/537.36'}

# with the request method, we make a GET request to the metacritic
resp = http.request('GET', url, retries=urllib3.util.Retry(), headers=header)

# print the status code of the response to verify the code
print("Status Verification:", resp.status)
print()

# retrieve the contents from raw bytes to text
datastring = str(resp.data, 'utf-8')

# print how much data we got from the site to verify the code
print("Total data found on the website:",len(datastring))
print()

# copied expression from last week's assignment  
title_pattern = "class=\"title\"><h3>(.*?)<\/h3><\/a>"
# find all non-overlapping matches of title_pattern in string as a list of string
movie_title = re.findall(title_pattern, datastring)

# copied expression from last week's assignment  
date_pattern = "<span>(.*?, 2019)<\/span>"
# find all non-overlapping matches of date_pattern in string as a list of string
movie_date = re.findall(date_pattern, datastring)

# copied expression from last week's assignment  
metascore_pattern = "class=\"metascore_w large movie positive\">(.*?)<\/div>\s*<\/a>\s*<\/div>\s*<span class=\"title numbered\">"
# find all non-overlapping matches of metascore_pattern in string as a list of string
movie_metascore = re.findall(metascore_pattern, datastring)

# copied expression from last week's assignment  
url_pattern = "<td class=\"clamp-image-wrap\">\s*<a href=\".*?\"><img src=\"(.*?)\" alt=\""
# find all non-overlapping matches of url_pattern in string as a list of string
movie_url = re.findall(url_pattern, datastring)

# copied expression from last week's assignment  
description_pattern = "class=\"summary\">\s*(.*)"
# find all non-overlapping matches of description_pattern in string as a list of string
movie_description = re.findall(description_pattern, datastring)
```

```python
# import pandas to create the table
import pandas as pd

# make data dictionary into a dataframe
movies = pd.DataFrame({"Title" : movie_title, "Date" : movie_date, "Metascore" : movie_metascore, "Thumbnail URL" : movie_url, "Description" : movie_description})

display(movies)
```
